```python
ds=['title', 'authors', 'original_publication_year', 'average_rating', 'ratings_count', 'small_image_url','books_count']
df = df[ds]
```

```python
# Filter Harry Potter books by J.K. Rowling
hp_books = df[ df['title'].str.contains('Harry Potter')]
hp_books
```

```python
#the most selling book
max_index = hp_books['books_count'].idxmax()  # Get index of the row with maximum value
max_value = df.loc[max_index, 'books_count']  # Get maximum value
title_value = df.loc[max_index, 'title']  # Get title 
print("Maximum value:", max_value)
print("Title :", title_value)

```
